The deterministic SIR model is given by the following system of ordinary differential equations:

$$
\frac{{\rm d} S}{{\rm d} t}  = - \beta S I, \quad 
\frac{{\rm d} I}{{\rm d} t}  = \beta S I - \mu I, \quad 
\frac{{\rm d} R}{{\rm d} t}  = \mu I.
$$

In these equations, $S$ is the number of susceptible individuals, $I$ is the number of infectious individuals, $R$ is the number of removed individuals (recovered and immune, or dead). $N$ is the total number of individuals, where $N = S+I+R$. By adding the three equations, it can be verified that $\frac{d N}{d t} = 0$, and so $N$ is constant. The variables $\beta$ and $\mu$ are model parameters: $\beta$ is the infection rate parameter, $\mu$ is the removal rate parameter (the reciprocal of the expected time for which an infected individual remains infectious).

As shown in the Epidemiology lecture, the basic reproduction number is given by
$ R_0 = \frac{\beta N}{\mu}. $

The Python code cells below compute numerical solutions to this model by calling the SciPy ODE solver odeint. 

Results are shown for specific parameter values: $ N=10^4$, $\beta = 2 \times 10^{-5}$ and $\mu = 0.1$, corresponding to $R_0 = 2$. The initial conditions are $S(0) = N - 3$, $I(0) = 3$, and $R(0) = 0$. So, we start with only 5 infectious individuals.

```python
# import libraries

import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint 
```

```python
# --- First define the RHS of ODE system --- #
#changed beta to match required starting value
beta = 2*10**-5
def SIR(y, t):
    # returns RHS of the SIR model
    S, I, R = y
    mu = 0.1
    
    N = S + I + R
    Sdot = -beta * S * I
    Idot = beta * S * I - mu * I
    Rdot = mu * I
    
    return Sdot, Idot, Rdot
```

```python
# --- Problem setup --- #

# set up time grid for the solution
tf = 100
Nsteps = 1000
t = np.linspace(0, tf, Nsteps+1)

# Set initial conditions and put into array y0
#changed S0 to match required starting value
S0 = 10**4 - 3
#changed I0 to match required starting value
I0 = 3
R0 = 0
y0 = np.array([S0, I0, R0])
```

```python
# --- Call the ODE solver and extract solutions --- #
  
y_sol = odeint(SIR, y0, t)

# extract S, I, R from the solution array y
S = y_sol[:,0]
I = y_sol[:,1]
R = y_sol[:,2]
```

```python
# --- Plot various interesting things --- #

# plot the time evolution of the S, I, R
plt.plot(t, S, label = "S")
plt.plot(t, I, lw = 4, label = "I")
plt.plot(t, R, label = "R")

#gets the index of the maximum value of I
max_index = np.argmax(I)
#plot this point as a black circle
plt.plot(t[max_index],I[max_index],'ko')


# labels etc
plt.xlabel("time", fontsize = 14)
plt.ylabel("S,I,R", fontsize = 14)
plt.title("SIR model", fontsize = 16)
plt.legend(fontsize = 14)
plt.show()        

# plot the evolution in the S,I phase plane
plt.plot(S, I)
#plot max value of I again
plt.plot(S[max_index],I[max_index],'ko')


# labels etc
plt.xlabel("S", fontsize = 14)
plt.ylabel("I", fontsize = 14)
plt.title("Phase plane for SIR model", fontsize = 16)

# put some markers to show initial and final states
plt.plot(S[0], I[0], 'rd', markersize = 10)
plt.plot(S[-1], I[-1], 'go', markersize = 10)
plt.show()

# --- Report the final values of S, I, R.  --- #

print("At the end of the simulation, the final values of S, I, and R are (rounded):")
print("S(t=tf) =", round(S[-1]))
print("I(t=tf) =", round(I[-1]))
print("R(t=tf) =", round(R[-1]))

print(f"the maximum value of I is {int(I[max_index])}.")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint 
import scipy.optimize as opt

def SIR(y, t):
    S, I, R = y
    beta = 2*(10**-5)
    mu = 0.1
    N = S + I + R
    Sdot = -beta * S * I
    Idot = beta * S * I - mu * I
    Rdot = mu * I
    return Sdot, Idot, Rdot
    
V = [1000, 2000, 3000]
tf = 500
Nsteps = 10000
t = np.linspace(0, tf, Nsteps+1)

I_plus_R = []
for i in range(len(V)):
    S0 = 10**4 - 5 - V[i]
    I0 = 5
    R0 = 0
    y0 = np.array([S0, I0, R0])
    y_sol = odeint(SIR, y0, t)
    S = y_sol[:,0]
    I = y_sol[:,1]
    R = y_sol[:,2]
    I_plus_R.append(I[-1] + R[-1])

# Set-Up for Final Size Equation
beta = 2*(10**-5)
mu = 0.1
N = 10**4

# Final Size Equation
def f(x): 
    return N - N*np.exp((-beta*x)/mu) - x 

# Finds result using Netwon-Raphson Method, plots using X's
final_size = []
estimates = [6500, 5250, 3600]
for j in range(len(V)):
    N = 10**4 - V[j]
    final_size.append(opt.newton(f, estimates[j]))
    plt.plot(V[j],final_size[j],marker='x',color='k',ms=6)
    print("Final Size Result when",V[j],"people are vaccinated:",opt.newton(f,estimates[j]))

# Plots I+R against Vaccinated Individuals
plt.plot(V,I_plus_R,marker='o',ms=2)
plt.xlabel("No. of Vaccinated People",fontsize=12)
plt.ylabel("I + R", fontsize=12)
plt.title("Infected + Recovered Individuals against Vaccinated Individuals 500 Days After", fontsize = 14)
plt.show()
```

```python
# libraries
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint 

# setting up SIR model - adding in being dependent on T
def SIR(y, t, T):
    S, I, R = y
    mu = 0.1
    if t < T:
        beta = 2*(10**-7)
    elif T <= t <= T + 30:
        beta = 1.5*(10**-7)
    else:
        beta = 2*(10**-7)
    
    N = S + I + R
    Sdot = -beta * S * I
    Idot = beta * S * I - mu * I
    Rdot = mu * I
    
    return Sdot, Idot, Rdot
    
# running SIR model for T = 30 days
tf = 300 # chose 300 days to allow for later T=200
Nsteps = 1000
t = np.linspace(0, tf, Nsteps+1)

S0 = 10**6 - 5
I0 = 5
R0 = 0
y0 = np.array([S0, I0, R0])

T_initial = 30
y_sol = odeint(SIR, y0, t, args=(T_initial,))
S = y_sol[:,0]
I = y_sol[:,1]
R = y_sol[:,2]

# plotting time series for S(t),I(t) and R(t) for T = 30 days
plt.figure(figsize=(10,6))
plt.plot(t, S, lw=1, label = "S")
plt.plot(t, I, lw = 2, label = "I")
plt.plot(t, R, lw=1, label = "R")

plt.xlabel("Time (days)", fontsize = 10)
plt.ylabel("S,I,R (# of individuals)", fontsize = 10)
plt.title("SIR model with social distancing (days 30 to 60)")
plt.legend()
plt.show() 
```

```python
# for loop to identify optimal value of T (minimises the maximum value of I)
max_I = np.zeros(201) 
max_I[0] = 10**9 # stops np.argmin(max_I) outputting max_I[0] which doesn't correspond to a T-value.

for T in range(1, 201):
    solution = odeint(SIR, y0, t, args=(T,))
    I_t = solution[:, 1]
    max_I[T] = np.max(I_t)
# max_I[0] is unused but then max_I[n] refers to T = n making indexing clearer
optimal_T = np.argmin(max_I)

# plotting maximum # of infected against T
plt.figure(figsize=(10, 6))
plt.plot(range(1, 201), max_I[1:], linewidth=2)

plt.xlabel('T (day social distancing starts)', fontsize=10)
plt.ylabel('Maximum number of infected individuals', fontsize=10)
plt.title('Maximum number of infected individuals against social distancing start time')
plt.show()

print("The value of T that minimises the maximum number of infected individuals the most is T =",optimal_T,".")
```

```python
# import libraries
import numpy as np 
import matplotlib.pyplot as plt
from scipy.integrate import odeint

# --- Problem setup --- #
SimulationNum=10000
# Parameter values
beta = 3.0*10**-4 
mu = 0.1

def ProblemSetup():
# Set up arrays to record event times, and values of S, I and R after after each event
    SVec = []
    IVec = []
    RVec = []
    tVec = []

# Set up an array to record pre-determined times and values of I at those times
    tVals = [i*0.5 for i in range(201)]
    IVals = [0 for i in range(201)]

# Set initial conditions and the initial time
    S = 10**3 - 1
    I = 1
    R = 0
    t = 0

    IVals[0] = I

# Record initial conditions in the S, I, R and t arrays
    SVec.append(S)
    IVec.append(I)
    RVec.append(R)
    tVec.append(t)
    return S, I, R, t, SVec, IVec, RVec, tVec, IVals, tVals

# --- While the outbreak is ongoing (I > 0), generate events --- #
def GenerateEvents(beta, mu, S, I, R, t, SVec, IVec, RVec, tVec, IVals, tVals):
    while I > 0:
        
        previous_time = t
        
        r1 = np.random.random()
        r2 = np.random.random()
        totalRate = beta*I*S + mu*I
        t = t - (1/totalRate)*np.log(r1) #finds time when next event will occur
        
        new_time = t
        
        for i in range(201):
            if previous_time < tVals[i] < new_time: 
                IVals[i] = I
        #tVals = 0,0.5,1,1.5,2,..., i*0.5, ..., 100
        #IVals = I1, I2, I3, I4, ..., i200, I200
        

        if r2 < beta*I*S/(totalRate): #find which event happens, infection or recovery, fixed mistake 2
            S = S - 1
            I = I + 1
        else: 
            I = I - 1
            R = R + 1
        
        SVec.append(S)
        IVec.append(I)
        RVec.append(R)
        tVec.append(t)
    return SVec, IVec, RVec, tVec, IVals, tVals

# --- generate simulations ---
recordedIVals=[]
recordedMVals=[]
for i in range(SimulationNum):
    S, I, R, t, SVec, IVec, RVec, tVec, IVals, tVals = ProblemSetup()
    SVec, IVec, RVec, tVec, IVals, tVals = GenerateEvents(beta, mu, S, I, R, t, SVec, IVec, RVec, tVec, IVals, tVals)
    recordedIVals.append(IVals) #store infected data
    recordedMVals.append(RVec[-1]) #store total recovery and infection data
 #   print(i)
recordedIVals = np.stack(recordedIVals, axis=0) 
```

## SIS model described in the paper

I chose to recreate the duration graph in the paper: figure 4b. The paper defines a severe epidemic as an outbreak that lasts for at least a certain number of days, and figure 4b plots the probability that an outbreak will persist for at least T days with different values for $R_0$ (the reproduction number).

first, we import some required libraries:

```python
from matplotlib.ticker import ScalarFormatter
from numba import jit
```

Next, we need to initialise the parameters. In the paper, the parameters used are $N=1000$, $I(0)=1$ and $S(0)=N-I(0)$. The paper states that $R_0$ is controlled using $\beta$ with the equation $R_0=\frac{\beta N}{\mu}$, which can be rearranged to give $\beta=\frac{R_0 \mu}{N}$, so we will use this to get our value for $\beta$. $\mu$ is also stated to be $0.1$ days.

the code for the simulation is essentially the same as that found in A5, except since we are using the SIS model, individuals who are no longer infected go back into the S population instead of moving into the R population. We also don't need to collect the S, I and t values over time as we only care about the final time.

```python
#set N and mu
N = 1000
mu=0.1
#this is a package that converts the following function to machine code before running it.
#this is not strictly necessary, but without it running all the simulations takes about 40 minutes.
@jit(nopython=True)
def run_simulation(R0,max_t):
    #the rest of the parameters rely on R0 or must be reset at the start of each simulation so they are set here.
    t = 0
    beta= R0 * mu / N
    I = 1
    S = N - I
    #we end the simulation either when there are no more infected individuals or the maximum time has been exceeded.
    while not (I <= 0 or t > max_t):
        #get 2 random values
        r1 = np.random.random()
        r2 = np.random.random()
        totalRate = beta*I*S + mu*I
        #calculate the next time an event occurs
        t = t - (1/totalRate)*np.log(r1)
        #decide what should happen on this event
        if r2 < beta*I*S/totalRate:
            S = S - 1
            I = I + 1
        else:
            I = I - 1
            S = S + 1
    return t
```

Next, we need to use this to plot the probability that the outbreak persists for T days. The paper states 10,000 simulations were run, so we do this by running 10,000 simulations, then calculate the proportion that have survived beyond the time we wish to plot.

```python
def get_graph(R0,col):
    max_t = 4000
    #the number of simulations to run
    n_sims = 10000
    t_values = []
    #run all the simulations
    for i in range(n_sims):
        t = run_simulation(R0,max_t)
        t_values.append(t)
    #the probability at each time
    probs = np.array([])
    times = np.array([])
    #evenly distribute 1000 points along a log scale
    points = np.logspace(np.log10(1), np.log10(max_t), 1000)
    for i in range(1000):
        #for each point, count up how many simulations survived beyond this time
        n = 0
        for j in range(n_sims):
            if t_values[j] >= points[i]:
                n += 1
        #divide by the total number of simulations to get the probability
        n /= n_sims
        probs = np.append(probs,n)
    #plot the horizontal line for R0 values greater than 1.
    if R0 > 1:
        plt.axhline(y=1-(1/R0),linestyle="--",color=col)
    #plot the graph
    plt.semilogx(points,probs,color=col,label=f"R0 = {R0}")
```

finally, we run this code for each $R_0$ value.

```python
plt.ylim(0,1)
plt.xlim(1,4000)
get_graph(0.8,'black')
get_graph(1.01,'blue')
get_graph(1.1,'red')
get_graph(1.3,'green')
get_graph(1.6,'purple')
get_graph(2,'grey')

#makes x axis display 1,10,100 etc. instead of 10^0, 10^1, 10^2 etc.
ax = plt.gca()
ax.xaxis.set_major_formatter(ScalarFormatter())

plt.xlabel("time (T days)")
plt.ylabel("probability persist for at least T days")

plt.legend()
plt.show()
```

## Research

One limitation of the SIR model is that it assumes constant rates of infection and recovery and that once an individual has recovered, they have permanent immunity. This limits the model when modelling various diseases such as COVID-19 because factors such as the implementation of government action such as social distancing, behavioural changes such as reduced social interaction and seasonal changes such as humidity affecting transmission, showing rates may vary over time $\text{[1]}$. Furthermore, for many diseases such as COVID-19, waning immunity means individuals in the recovered population may gradually are reintroduced into the susceptible population. Therefore, to account for these two factors, the constant rates can be replaced by time-dependent functions as well as extending the SIR model into an SIRS model to better model for diseases where reinfection may occur.
Therefore, our differential equations:

$$
\frac{{\rm d} S}{{\rm d} t}  = - \beta S I, \quad 
\frac{{\rm d} I}{{\rm d} t}  = \beta S I - \mu I, \quad 
\frac{{\rm d} R}{{\rm d} t}  = \mu I.
$$

would become:

$$
\frac{{\rm d} S}{{\rm d} t}  = - \beta(t) S I, \quad 
\frac{{\rm d} I}{{\rm d} t}  = \beta(t) S I - \mu(t) I, \quad 
\frac{{\rm d} R}{{\rm d} t}  = \mu(t) I.
$$

where $\beta(t)$ and $\mu(t)$ represent the time-dependent infection rate and removal rate parameters respectively.

Furthermore, we introduce a third model parameter representing the loss of immunity, moving individuals from the recovered population back into the susceptible population modifying two of the differential equations further:

$$
\frac{{\rm d} S}{{\rm d} t}  = \gamma(t) R - \beta(t) S I, \quad 
\frac{{\rm d} I}{{\rm d} t}  = \beta(t) S I - \mu(t) I, \quad 
\frac{{\rm d} R}{{\rm d} t}  = \mu(t) I - \gamma(t) R.
$$

where $\gamma(t)$ represents the time-dependent parameter for the rate of loss of immunity.

If reinfection does not occur, we can set $\gamma(t)$ = 0  for all values of t to reduce the system of differential equations back to the original SIR model.

Another limitation of the SIR model, particularly when modelling diseases such as Measles and Influenza, is that the model cannot account for different transmission rates between age groups. On average, one would expect young people to have higher overall contact rates than elderly people, and for individuals to tend to interact predominantly with others within their own age group due to school, workplaces, and social settings. In order to increase the accuracy of our model, age-structured modelling can be introduced to make the SIR more applicable to real-world disease transmission. These models divide the population into age categories, where parameters like infection and removal rate are distinct as a result of varying contact rates. To implement this into our model, we would introduce a ‘contact matrix’ $A=\begin{pmatrix}a_{ij}\end{pmatrix}$ where “the constant $a_{ij}$ represents the contact rate of age group *j* with age group *i*” $\text{[2]}$. For example, our current differential equation $$\frac{dS}{dt} = -\beta SI$$ would become $$\frac{dS_i}{dt} = -S_i \sum_{0 < j \leq n} a_{ij} I_j \quad \text{[2]}$$ where *i* indexes the age category we are modelling, and *n* refers to how many age categories we are including in our model. Implementing contact matrices hen modelling a disease prevalent in a specific age group would greatly improve our model. When modelling measles or chicken-pox, it is not realistic to assume that children will interact with elderly individuals as much as they would with other children.

Together, these two extensions of the SIR models would improve its applicability to real-life scenarios, and drastically decrease the number of assumptions we make when modelling diseases.
